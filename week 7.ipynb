```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Given data
study_hours = np.array([2, 4, 6, 8, 10])
exam_scores = np.array([50, 60, 68, 85, 95])
sleep_hours = np.array([8, 7, 5.5, 5.5, 4])

# Scatter plots
plt.figure(figsize=(12, 4))

# Study Hours vs Exam Scores
plt.subplot(1, 3, 1)
plt.scatter(study_hours, exam_scores, color='blue')
plt.title('Study Hours vs. Exam Scores')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')

# Sleep Hours vs Exam Scores
plt.subplot(1, 3, 2)
plt.scatter(sleep_hours, exam_scores, color='green')
plt.title('Sleep Hours vs. Exam Scores')
plt.xlabel('Sleep Hours')
plt.ylabel('Exam Scores')

# Study Hours vs Sleep Hours
plt.subplot(1, 3, 3)
plt.scatter(study_hours, sleep_hours, color='red')
plt.title('Study Hours vs. Sleep Hours')
plt.xlabel('Study Hours')
plt.ylabel('Sleep Hours')

# Display the scatter plots
plt.tight_layout()
plt.show()

# Pearson Correlation Coefficients
r_study_exam, _ = pearsonr(study_hours, exam_scores)
r_sleep_exam, _ = pearsonr(sleep_hours, exam_scores)
r_study_sleep, _ = pearsonr(study_hours, sleep_hours)

# Print the Pearson correlation coefficients
print(f"Pearson Correlation (Study Hours vs. Exam Scores): {r_study_exam}")
print(f"Pearson Correlation (Sleep Hours vs. Exam Scores): {r_sleep_exam}")
print(f"Pearson Correlation (Study Hours vs. Sleep Hours): {r_study_sleep}")
```
